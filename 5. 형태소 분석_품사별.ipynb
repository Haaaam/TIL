```python
from konlpy.tag import Okt
okt=Okt()
```

```python
okt.tagset
```

```python
text=open('./data/2021 신년사.txt','r').read()
word_txt=okt.pos(text)
display(word_txt)
```

```python
형용사=[]
명사=[]
부사=[]

for word in word_txt:
    if word[1]=='Adjective':
        형용사.append(word[0])
    elif word[1]=='Adverb':
        부사.append(word[0])
    elif word[1]=='Noun':
        명사.append(word[0])
print('[명사]', 명사)
print('\n\n[부사]', 부사)
print('\n\n[형용사]', 형용사)
```

```python
from wordcloud import WordCloud, STOPWORDS
from collections import Counter
import matplotlib.pyplot as plt
import nltk
```

```python
text_ko=okt.nouns(text)
```

```python
ko_txt=nltk.Text(text_ko)
ko_txt
```

```python
stop_words=['.', ',', '(',')', "'", '%', '-', 'X', ').', '×','의','자','에','안','번','수',
            '호','을','이','다','만','로','가','를', '곳','여러분', '부','전', '것','일 ',
            '등','위','못', '데', '나', '말','해', '낼','포', '일', '년', '속', '우리', '국민']
text_ko=[each_w for each_w in ko_txt if each_w not in stop_words]
text_ko
```

```python
counts=Counter(text_ko)
wordcloud=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                   stopwords=stop_words,
                   background_color='white',
                   width=1200,height=1000).generate_from_frequencies(counts)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
cnt_txt=counts.most_common()
cnt_txt
```

```python
import pandas as pd

df1=pd.DataFrame(cnt_txt,columns=['단어','빈도수'])
df2=df1[df1['빈도수']>=10]
df2
```

```python
import matplotlib.font_manager as fm

fontprop = fm.FontProperties(fname='C:/Windows/Fonts/HANDotum.ttf', size=15).get_name()
plt.rc('font', family=fontprop)

plt.pie(df2['빈도수'], labels=df2['단어'], autopct='%.1f%%')
plt.show()
```
