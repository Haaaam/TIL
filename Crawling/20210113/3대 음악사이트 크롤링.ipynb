```python
from selenium import webdriver
from datetime import datetime
from bs4 import BeautifulSoup as bs
import time
import requests
import numpy as np
import pandas as pd
```

```python
def bugs():
    url='https://music.bugs.co.kr/chart'
    html=requests.get(url)
    soup=bs(html.content, 'html.parser')
    
    rank_tmps=soup.find_all('div',class_='ranking')#순위
    p_tmps=soup.find_all('p',class_='title')#곡명
    artist_tmps=soup.find_all('p',class_='artist')#아티스트
    alb_tmps=soup.find_all('a',class_='album')#앨범
    
    bugs_rank=[]
    for i in range(len(rank_tmps)):
    
        rank=rank_tmps[i].find('strong').text
        p=p_tmps[i].find('a').text
        artist=artist_tmps[i].find('a').text
        alb=alb_tmps[i+1]['title']
        alb_link=alb_tmps[i+1]['href']

        bugs_rank.append([rank,p,artist,alb,alb_link])
    
    df1=pd.DataFrame(bugs_rank,columns=['순위','곡명','아티스트','앨범','앨범링크'])
    df1.to_csv('./data/20210113_Bugs! Rank100.csv',index=False)
    display(df1.head(5))
```

```python
def vibe():
    driver=webdriver.Chrome('d:/py_data/chromedriver.exe')
    url='https://vibe.naver.com/chart'

    driver.get(url)
    driver.implicitly_wait(5)
    
    driver.find_element_by_xpath('//*[@id="content"]/div[1]/div/h3/a').click()

    html=driver.page_source
    soup=bs(html,"html.parser")
    
    #rank
    rank_lst=soup.find_all('td',class_='rank')
    #곡명
    sing_lst=soup.find_all('td',class_='song')
    #가수
    artist_lst=soup.find_all('td',class_="artist")
    #앨범
    link_lst=soup.find_all('td',class_="album")

    vibe_lst=[]

    u='https://vibe.naver.com'

    for i in range(len(rank_lst)):
        rank=rank_lst[i].find('span').text
        sing=sing_lst[i].find('a').text
        artist=artist_lst[i].find('span').text
        alb=link_lst[i].find('a').text
        alb_link=str(u)+link_lst[i].find('a')['href']

        vibe_lst.append([rank,sing,artist,alb,alb_link])
        
    driver.close()    
    today=datetime.today()
    today=today.strftime("%Y%m%d")
    
    df2=pd.DataFrame(vibe_lst, columns=['순위', '곡명', '아티스트', '앨범', '앨범링크'])    
    df2.to_csv('./data/'+str(today)+'_Vibe Rank100.csv', index=False)
    display(df2.head(5))
    
```

```python
def melon():
    driver=webdriver.Chrome('d:/py_data/chromedriver.exe')

    url='https://www.melon.com/chart/index.htm'

    driver.get(url)
    driver.implicitly_wait(5)

    html=driver.page_source
    soup=bs(html,"html.parser")
    
    
    #곡명
    sing_lst=soup.find_all('div',class_='ellipsis rank01')
    #가수명
    artist_lst=soup.find_all('div',class_='ellipsis rank02')
    #앨범명
    alb_lst=soup.find_all('div',class_='ellipsis rank03')


    melon_lst=[]
    for i in range(len(sing_lst)):
        sing=sing_lst[i].find('a').text
        artist=artist_lst[i].find('span',class_='checkEllipsis').text
        alb=alb_lst[i].find('a').text
        alb_href=alb_lst[i].find('a')['href']
        melon_lst.append([i+1,sing,artist,alb,alb_href])
        
    driver.close()   
    today=datetime.today()
    today=today.strftime("%Y%m%d")
    
    df3=pd.DataFrame(melon_lst,columns=['순위','곡명','아티스트','앨범','앨범링크'])
    df3.to_csv('./data/'+str(today)+'_Melon Rank100.csv',index=False)
    display(df3.head(5))
```

```python
#Main

bugs()
print('벅스뮤직 완료')
vibe()
print('vibe 완료')
melon()
print('멜론뮤직 완료')
```
