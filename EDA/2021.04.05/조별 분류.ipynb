```python
import pandas as pd
import os

tt=os.getcwd()
file_path=os.path.join(tt,'c:/py_data/조배정.txt')
f=open(file_path,encoding='utf-8')
group_num=[]

df_group=pd.DataFrame(columns=['번호','이름','그룹'])
for line in f.readlines():
    if len(line)>2:
        if 'PC' in line:
            line=line.split('님이')[0]
            line=line.split('PC')[0]
        else:
            line=line.split('님이')[0]
        num=line[:2]
        name=line[3:]
        df_group=df_group.append(pd.DataFrame([[num,name]],columns=['번호','이름']))
    else:
        group_num.append(int(line[:1]))

df_group['그룹']=group_num
df_group
```

```python
#데이터 프레임 분할하기
for i in range(1,max(df_group['그룹'])+1):
    df_i=df_group.loc[df_group['그룹']==i]
    display(df_i)
```
