### 데이터 결측치 시각화

```python
#결측치 시각화
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('1.csv')
df=df.isnull()
```

```python
plt.rcParams['font.family']='Malgun Gothic'
sns.heatmap(df)
```

```python
tmp=df.isna().sum()
dfNa=tmp[tmp>0]
dfNa
```

```python
naCnt=tmp[0]
if len(df)*0.1>=naCnt:
    print('제거')
if len(df)*0.1<naCnt<len(df)*0.2:
    print("임의로 대체")
else:
    print("예측값으로 대체")
```
