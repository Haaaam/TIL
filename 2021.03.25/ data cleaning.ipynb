```python
import pandas as pd
df=pd.read_csv('./3-1.csv',index_col=0)

df.plot(kind='box',y='abv')
df['abv'].describe()
```

```python
df['abv'].describe()[4] #25%Q1의 값을 확인할 수 있음
```

```python
q1=df['abv'].quantile(0.25)
q3=df['abv'].quantile(0.75)

iqr=q3-q1

#이상치 확인
re=(df['abv']<q1-iqr*1.5) | (df['abv']>q3*iqr*1.5)
df[re] #re의 true인 값만 출력함
```

```python
#2250번행에 0.055
#이상치를 확인함

df['abv'][2250]=0.055
```
