```python
f=open('./data/2021 신년사.txt','r')
text=f.read()
f.close()
```

#### Kkma Class
- analyze(text):각 토론에 대해 다양한 형태로 반환, 해당 분석기는 사전검색(차트), 분류되지 않은 용어 두 부분으로 분할 구성되어 있다.
- morphs(text):텍스트에서 형태소로 반환한다.
- nouns(text): 텍스트에서 명사를 반환한다.
- pos(text): 텍스트에서 품사 정보를 부착하여 반화한다.

```python
from konlpy.tag import Okt
okt=Okt()
nouns_txt=okt.nouns(text)
nouns_txt
```

```python
from collections import Counter
cnt=Counter(nouns_txt)
cnt
```

```python
rank_txt=cnt.most_common() #most_common(): 내림차순으로 정렬하는 함수
rank_txt
```

```python
rank_txt[0][1]
```

```python
for (keyword,num) in rank_txt:
    if num<10:
        print(keyword,num)
        rank_txt.remove((keyword,num))
```

```python
rank_txt.remove(('발견',1))
rank_txt
```

```python
dic_rank=dict(rank_txt)
dic_rank
```
