```python
import torch
```

```python
#변수 선언(+데이터 입력)
x=torch.ones(2,2,requires_grad=True)  
```

```python
# 모델 내 연산 예측값 산출
y=x+1
print(y)
```

```python
# 손실함수 계산
z=2*y**2
res=z.mean()
```

```python
# 손실 산출
print(res)
```

```python
#d(res)/dx_i=x_i+1
# res=(z_1+..+z_4)/4
# z_i=2*y_i**2
# z_i=2(x_i+1)**2
res.backward()
```

```python
print(x.grad)
```
