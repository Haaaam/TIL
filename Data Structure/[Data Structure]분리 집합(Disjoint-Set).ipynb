#### 분리 집합이란?

- 서로 겹치는 원소가 없는 집합(서로소 집합)들을 관리하는 자료 구조

- 분리 집합에는 두가지 연산이 존재한다.

  1) Find : 원소 x가 속한 집합을 찾는 연산으로 어떤 두 원소가 같은 집합에 속하는지 알 수 있다.

  2) Union: 두 집합을 합치는 연산

- 'Union-Find'라고도 부른다.

#### 분리 집합의 구현



- 트리구조는 부모노드가 자식 노드를 가리키지만 분리 집합은 자식 노드가 부모 노드를 가리키는 형태이다. 

   (부모-자식 관계를 가리키는 방향이 반대)

- 가장 효율적인 트리 구조를 사용해서 구현한다.









예를 들기 위해 만들어 놓은 위의 분리 집합을 보면 4 와 2가 각각의 집합의 대표 노드(root node)가 됩니다.

즉, 트리 하나를 각각 한개씩의 집합으로 보면 되는 것입니다.

이제 분리집합을 만들었으니 연산을 해봅시다.

앞서 분리 집합의 정의에서 말했듯이 분리집합의 연산에는 Find와 Union 연산이 있습니다.



Find 연산: 해당 원소 노드가 속해 있는 집합, 즉 루트 노드를 찾는 연산



예를 들어 왼쪽 분리 집합의 5번 노드에서 find 연산을 수행한다면 해당 집합의 root node인 4를 return 값으로 갖게 됩니다.

8번 노드에서 find 연산을 수행해도 같은 결과 값인 4를 return 값으로 갖게 됩니다.

이것으로 봤을때 8번 노드와 5번 노드가 같은 집합에 속해 있다는 것을 알 수 있습니다.


Union 연산:  두 집합을 하나로 합치는 연산

- 집합 A,B가 있는데 이 두 집합에 Union 연산을 한다고 하자. 이때, 교집합이 되는 것이 아닌 집합 A(또는 B)의 부모를 집합 B(또는 A)로 설정하는 것이다.



  위의 두 분리 집합을 하나로 합친다고 하면 아래 그림처럼 2의 부모 노드를 부모 노드 4에 연결하여 하나의 트리로 만들어 줍니다. 그 결과, 하나의 집합이 되는 것을 확인할 수 있습니다.
