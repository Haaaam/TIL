```python
import time
import os
import pandas as pd

fdir="./topics/"
file_lst=os.listdir(fdir)
files=[]
opinion_text=[]
for i in file_lst:
    df=pd.read_table(fdir+i,header=0,encoding='latin1')
    fname=i.split('.')[0]
    files.append(fname)
    opinion_text.append(df.to_string()) #to_string()이용시, 넘버번호가 자동으로 생성됨
df_document=pd.DataFrame({'filename':files,'opinion_txt':opinion_text})   
df_document.to_csv('movies_opinion.csv')
```

```python
import pandas as pd
import glob ,os

path = './topics'
all_files = glob.glob(os.path.join(path, "*.data"))     
filename_list = []
opinion_text = []


for file_ in all_files:
    df = pd.read_table(file_,index_col=None, header=0,encoding='latin1')
    filename_ = file_.split('\\')[-1]
    filename = filename_.split('.')[0]
    filename_list.append(filename)
    opinion_text.append(df.to_string())

document_df = pd.DataFrame({'filename':filename_list, 'opinion_text':opinion_text})
```
