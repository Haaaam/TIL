### 이미지 각도 변경
- pic1, pic2, pic3를 이미지를 읽어와서 이미지 각도를 90도 회전
- 기존 파일명 앞에 rot90_을 붙여서 저장(예: rot90_pic1.jpg)


```python
#tmp.rotate(회전각도) : 이미지를 회전각도 만큼 회전, 
#저장변수=>tmpRo
from PIL import Image
fileName='pic1'
tmp=Image.open('./picture/'+fileName+'.jpg')
tmpRo=tmp.rotate(90)
tmpRo.save('./img/'+'rot90_'+fileName+'.jpg')
```

### [미션]

### 함수
- 미리 작성해놓은 논리식을 함수라합니다.
- 함수는 함수명()로 작성합니다.
- 사용자가 사용하는 모듈(패키지, 라이브러리)은 함수의 집합입니다.
- 사용자가 직접 작성하는 사용자 정의함수는 파이썬에서는 def 함수명으로 작성합니다.

```python
#사용자 정희 함수, 실행 함수
def picSize(fileName,size):
    tmp=Image.open('./picture/'+fileName+'.jpg')
    tmpResize=tmp.resize((size,size))
    tmpResize.save('./picture/'+fileName+'_'+str(size)+'.jpg')
```

```python
from PIL import Image

picSize('pic1',900)
picSize('pic2',300)
picSize('pic3',600)
```

```python
#사용자 정희 함수, 실행 함수
def pSize(fileName,size):
    tmp=Image.open('./picture/'+fileName)
    tmpResize=tmp.resize((size,size))
    tmpResize.save('./img/'+str(size)+'_'+fileName)
```

```python
from PIL import Image
import os              #시스템 자체

fname=os.listdir('./picture/')
print('원본', fname)

for i in range(len(fname)):
    pSize(fname[i],300)
    
fileName=os.listdir('./img/')
print('변경',fileName)
```

```python
from PIL import Image
import pandas as pd
import os
```

```python
fileName=os.listdir('./picture/')
dic1=[]
for i in fileName:
    img=Image.open('./picture/'+i)
    dic1.append([i,img.format,img.width,img.height])
#dic1.append([fileName[0], img.format,img.width,img.height])
#img=Image.open('./picture/'+fileName[1])
#dic1.append([fileName[1], img.format,img.width,img.height])

print(dic1)
```

```python
# 모듈 연결
from PIL import Image
import pandas as pd
import os

fileName=os.listdir('./picture/')
dic2=[]
for i in range(len(fileName)):
    img=Image.open('./picture/'+fileName[i])
    dic2.append({'파일명':fileName[i],'파일형식':img.format,'너비':img.width,'높이':img.height})

#print(dic2)

df1=pd.DataFrame(dic2)
df2=pd.DataFrame(dic2,columns=['파일명','너비','높이','파일형식'])

display(df1)
display(df2)
```

```python
#csv 저장
df2.to_csv('./img/파일정보.csv') #, 기본값: encoding='utf-8'

#인코딩 :https://freestrokes.tistory.com/71
```
