### 파이썬 워드클라우드 기본


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family']
tmp='강아지 산책 강아지 목욕 강아지 미용 강아지 쇼핑 친구와 저녁 먹음 가족과 점심 먹음 혼자 저녁 먹음 친구와 쇼핑'
wc=WordCloud(font_path='./NanumGothic.ttf',
             background_color='black',
             width=1000,height=1000,
             max_words=10,
             max_font_size=200)
wordFre=wc.generate(tmp)
print(wordFre)
display(wordFre.words_)
```

### 워드클라우드_파일

```python
#이미지 출력
plt.figure(figsize=(12,12))
plt.imshow(wordFre,interpolation='bilinear')
plt.axis('off')
plt.show()
```

### 2021 대통령 신년사 워드클라우드


- read():전체 읽어오기
- readline():행단위 읽어오기
- readlines():한개의 행을 리스트 형식으로 읽어오기


```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.rcParams['font.family']='NanumGothic'
```

```python
f=open('./data/2021 신년사.txt','r')
lines=f.read()
f.close()
print(lines)
```

```python
wc=WordCloud(font_path='./NanumGothic.ttf',
            background_color='black')
wordFre=wc.generate(lines)
print(wordFre)
display(wordFre.words_)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wordFre,interpolation='bilinear')
plt.axis('off')
plt.show()
```
