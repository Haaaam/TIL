#### yield란?
- 생산하다라는 뜻과 함께 양보하다라는 뜻을 가지고 있다.
- yield를 사용하면 값을 함수 바깥으로 전달하면서 코드 실행을 함수 바깥에 양보한다.


```python
# yield 동작 과정 예시
def number_generator():
    yield 0
    yield 1
    yield 2
    
g=number_generator()

a=next(g) #yield를 사용하여 함수 바깥으로 전달한 값은 next의 반환값으로 다시 나온다.
print(a)

b=next(g)
print(b)

c=next(g)
print(c)

```
